```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
data = pd.read_csv("Accidents_US.csv")   # change name if needed

# Display dataset
data
```

```python
data['Start_Time'] = pd.to_datetime(data['Start_Time'], errors='coerce')
data.dropna(subset=['Start_Time'], inplace=True)

# Correct accessor is .dt
data['Hour'] = data['Start_Time'].dt.hour
data['Day'] = data['Start_Time'].dt.day_name()

```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Hour', data=data)
plt.title('Accidents by Time of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Accidents')
plt.show()

```

```python
top_weather = data['Weather_Condition'].value_counts().head(10)

plt.figure(figsize=(10,5))
top_weather.plot(kind='bar')
plt.title('Top 10 Weather Conditions During Accidents')
plt.xlabel('Weather Condition')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Severity', data=data)
plt.title('Accident Severity Distribution')
plt.xlabel('Severity Level')
plt.ylabel('Count')
plt.show()
```
